```python
import tensorflow_decision_forests as tfdf
import pandas as pd
import numpy as np
```

```python
ds_train = pd.read_csv('train.csv')
ds_test = pd.read_csv('test.csv')
```

```python
print (ds_train.dtypes)
```

```python
object_columns = ds_train.select_dtypes(include=['object']).columns
print(object_columns)

object_columns_test = ds_test.select_dtypes(include=['object']).columns
print(object_columns_test)
```

```python
#len(object_columns)
```

```python
col_dummy = []
for column in ds_train.columns:
    if ds_train[column].nunique()==2:
        col_dummy.append(column)
print(col_dummy)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in col_dummy:
    ds_train[column] = le.fit_transform(ds_train[column])
    ds_test[column] = le.fit_transform(ds_test[column])
```

```python
#print(ds_train)
```

```python
object_columns = [item for item in object_columns if item not in col_dummy]
```

```python
print(object_columns)
```

```python
le = LabelEncoder()
for column in object_columns:
    ds_train[column] = le.fit_transform(ds_train[column])
    ds_test[column] = le.fit_transform(ds_test[column])
```

```python
ds_train
```

```python
columns_with_missing = ds_train.columns[ds_train.isnull().any()]
print(columns_with_missing)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(ds_train[columns_with_missing])
ds_train[columns_with_missing] = imputer.transform(ds_train[columns_with_missing]) 
```

```python
ds_train
```

```python
X = ds_train.iloc[:, :-1].values
y = ds_train.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
columns_with_missing_test = ds_test.columns[ds_test.isnull().any()]
print(columns_with_missing_test)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(ds_test[columns_with_missing_test])
ds_test[columns_with_missing_test] = imputer.transform(ds_test[columns_with_missing_test]) 
```

```python
X_test_to_know = ds_test.iloc[:, :].values
```

```python
y_pred_final = regressor.predict(X_test_to_know)
np.set_printoptions(precision=2)
print(y_pred_final)
```

```python
X_test_to_know[:, 0]
```

```python
submission = pd.DataFrame(data = {'Id': (X_test_to_know[:, 0]).astype(int),
                                'SalePrice': y_pred_final})

submission.to_csv('predicted_prices_submission.csv', index=False)
```
